🧪 실습: 타이타닉 생존자 데이터 분석

📦 [문제 0] 데이터 불러오기

✅ 문제

titanic.csv 파일을 읽어서 DataFrame으로 저장하고
상위 5개 데이터를 출력하시오.

```python
import pandas as pd

df = pd.read_csv("/home/titanic.csv")

print(df.head())
```

🧪 [문제 1] 기본 통계 확인

✅ 문제

1. 승객의 평균 나이를 구하시오.

2. 최고 나이를 구하시오.

3. 생존자의 비율을 구하시오.

```python
print("평균 나이:", df["Age"].mean())
print("최고 나이:", df["Age"].max())
print("생존 비율:", df["Survived"].mean())
```

🧪 [문제 2] 가설 설정

👉 “여성이 남성보다 생존률이 높다”

✅ 문제

성별(Sex) 기준으로 생존률을 구하시오.

```python
gender_survival = df.groupby("Sex")["Survived"].mean()
print(gender_survival)
```

🧪 [문제 3] 시각화 (성별 생존률)

✅ 문제

성별 생존률을 막대그래프로 나타내시오.

```python
import matplotlib.pyplot as plt

plt.bar(gender_survival.index, gender_survival.values)

plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")

plt.show()
```

🧪 [문제 4] 또 다른 가설

👉 “상위 클래스일수록 생존률이 높다”

✅ 문제

Pclass(좌석 등급)별 생존률을 구하시오.

```python
class_survival = df.groupby("Pclass")["Survived"].mean()
print(class_survival)
```

🧪 [문제 5] 시각화 (클래스별 생존률)

✅ 문제

클래스 별 생존률을 선 그래프로 나타내시오.

```python
class_survival = df.groupby("Pclass")["Survived"].mean()

plt.plot(class_survival.index, class_survival.values, marker='o')

plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Class")

plt.show()
```

🧪 [문제 7] 나이와 생존 관계

👉 “어린 사람이 더 많이 생존했을 것이다”

✅ 문제

20세 미만과 20세 이상 그룹의 생존률을 출력하시오.

```python
df["AgeGroup"] = "Adult"
df.loc[df["Age"] < 20, "AgeGroup"] = "Young"

age_survival = df.groupby("AgeGroup")["Survived"].mean()
print(age_survival)
```

🧪 [문제 8] 나이와 좌석 등급 관계

✅ 문제

승객의 나이(Age)와 좌석 등급(Pclass)의 관계를 시각화하시오.

요구사항

1. x축: 좌석 등급(Pclass)

2. y축: 나이(Age)

3. 막대그래프를 그리시오.

4. 그래프의 제목과 축 이름을 설정하시오.



```python
avg_age = df.groupby("Pclass")["Age"].mean()

plt.bar(avg_age.index, avg_age.values)

plt.xlabel("Pclass")
plt.ylabel("Average Age")
plt.title("Average Age by Class")

plt.show()
```
