📦 [문제 0] 데이터 불러오기

✅ 문제

titanic.csv 파일을 읽어서 DataFrame으로 저장하고
상위 5개 데이터를 출력하시오.

```python
import pandas as pd

df = pd.read_csv("/content/titanic.csv")

print(df.head())
```

🧪 [문제 1] 기본 통계 확인

✅ 문제

1. 승객의 평균 나이를 구하시오.

2. 최고 나이를 구하시오.

3. 생존 여부(Survived)의 평균을 구하여 생존률을 계산하시오

```python
print("평균 나이:", df["Age"].mean())
print("최고 나이:", df["Age"].max())
print("생존 비율:", df["Survived"].mean())
```

🧪 [문제 2] 가설 설정

👉 “여성이 남성보다 생존률이 높다”

✅ 문제

성별(Sex) 기준으로 생존률을 구하시오.

```python
gender_survival = df.groupby("Sex")["Survived"].mean()
print(gender_survival)
```

🧪 [문제 3] 시각화 (성별 생존률)

✅ 문제

성별 생존률을 막대그래프로 나타내시오.

```python
import matplotlib.pyplot as plt

gender_survival.plot(kind='bar')

plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")

plt.show()
```

🧪 [문제 4] 또 다른 가설

👉 “상위 클래스일수록 생존률이 높다”

✅ 문제

Pclass(좌석 등급)별 생존률을 구하시오.

```python
class_survival = df.groupby("Pclass")["Survived"].mean()
print(class_survival)
```

🧪 [문제 5] 시각화 (클래스별 생존률)

✅ 문제

좌석 등급 별 생존률을 선 그래프로 나타내시오.

```python
class_survival = df.groupby("Pclass")["Survived"].mean()

class_survival.plot(kind='line', marker='o')

plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Class")

plt.show()
```

🧪 [문제 6] 나이와 생존 관계

👉 “어린 사람이 더 많이 생존했을 것이다”

✅ 문제

🧪 [문제 7] 나이와 좌석 등급 관계

✅ 문제

좌석 등급(Pclass)별 평균 나이를 구하고 시각화하시오

요구사항

🧪 [문제 8] 조건 만족 데이터 탐색

✅ 문제

1. 여성
2. 1등석
3. 생존한 사람

🧪 [문제 9] 생존자와 사망자의 나이 분포 시각화

✅ 문제

성별(Sex)과 좌석 등급(Pclass)에 따른 생존률을 그래프로 비교하시오

요구사항

1. 성별(Sex)과 좌석등급(Pclass)으로 그룹을 나누시오

2. 각 그룹의 생존률을 계산하시오

3. 결과를 하나의 선 그래프로 나타내시오  
 (x축: Pclass, y축: Survival Rate)

```python
group_survival = df.groupby(["Pclass", "Sex"])["Survived"].mean()
group_survival = group_survival.unstack()

group_survival.plot(kind='line', marker='o')
plt.xlabel("Pclass")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Class and Sex")

plt.legend()
plt.show()
```

📦 문제 11. NumPy로 그룹 분류 (AgeGroup2)

✅ 문제

다음 기준으로 새로운 열 AgeGroup2를 생성하시오.

*   18 미만 → Child
*   18~59 → Adult
*   60 이상 → Senior

👉 NumPy를 활용하시오.

```python
age_np = df["Age"].values

age_group2 = np.where(age_np < 18, "Child",
               np.where(age_np < 60, "Adult", "Senior"))

df["AgeGroup2"] = age_group2
print(df[["Age", "AgeGroup2"]].head())
```

📦 문제 12. NumPy 기반 비교

✅ 문제
1. 일부 Age 값에 결측 값이 있다. 해당 NAN 값들을 제거하시오. (dropna() 활용)

2. 남성과 여성의 나이를 NumPy 배열로 분리하여 각각 평균 나이를 구하시오.

3. 전체 승객의 나이에 대한 표준편차를 구하시오.

```python
male_age = df[df["Sex"] == "male"]["Age"].dropna().values
female_age = df[df["Sex"] == "female"]["Age"].dropna().values

print("남성 평균 나이:", male_age.mean())
print("여성 평균 나이:", female_age.mean())
```

📦 문제 14. NumPy로 이상치 탐지 (Outlier)

✅ 문제

나이(Age) 데이터를 NumPy 배열로 변환한 후, 평균 ± 2×표준편차 범위를 벗어난 값을 이상치로 정의한다.

👉 이상치에 해당하는 나이를 출력하시오

```python
import numpy as np

age_np = df["Age"].dropna().values

mean = age_np.mean()
std = age_np.std()

lower = mean - 2 * std
upper = mean + 2 * std

outliers = age_np[(age_np < lower) | (age_np > upper)]

print("이상치 나이:", outliers)
```
