🧪 실습: 쇼핑 데이터 분석

📦 [문제 0] 데이터 불러오기

✅ 문제

shopping.csv 파일을 읽어서 DataFrame으로 저장하고
상위 5개 데이터를 출력하시오.

```python
import pandas as pd

df = pd.read_csv("/content/shopping.csv")

print(df.head())
```

🧪 [문제 1] 구매 금액 계산

✅ 문제

1. 각 주문의 총 구매 금액을 계산하시오.

* 총 구매 금액 = UnitPrice × Quantity

2. 계산 결과를 새로운 열 TotalPrice로 추가하시오
3. 전체 주문의 평균 구매 금액을 구하시오




```python
# 총 구매 금액 계산
df["TotalPrice"] = df["UnitPrice"] * df["Quantity"]

# 평균 구매 금액
print("평균 구매 금액:", df["TotalPrice"].mean())
```

🧪 [문제 2] 카테고리별 매출 분석

✅ 문제

1. Category 별로 평균 구매 금액(TotalPrice)을 구하시오.
2. Category 별 평균 구매 금액을 선 그래프로 시각화하시오
*   x축: Category, y축: TotalPrice





```python
category_sales = df.groupby("Category")["TotalPrice"].mean()
print(category_sales)
```

```python
import matplotlib.pyplot as plt

category_sales.plot()

plt.xlabel("Category")
plt.ylabel("Average Sales")
plt.title("Average Sales by Category")

plt.show()
```

🧪 [문제 3] 지역별 소비 분석

✅ 문제

1. Region 별 평균 구매 금액을 구하시오.
2. Region 별 평균 구매 금액을 막대그래프로 시각화하시오.
*   x축: Region, y축: Total Price




```python
region_sales = df.groupby("Region")["TotalPrice"].mean()
print(region_sales)
```

```python

region_sales.plot(kind='bar')

plt.xlabel("Region")
plt.ylabel("Average Sales")
plt.title("Average Sales by Region")

plt.show()
```

🧪 [문제 4] 구매 금액 분포 분석

✅ 문제

1. TotalPrice 값을 이용하여 히스토그램으로 나타내시오.
*   x축: TotalPrice, y축: count
*   그래프 제목: Total Purchase Distribution



```python
df[["TotalPrice"]].plot(kind='hist')

plt.xlabel("TotalPrice")
plt.ylabel("Count")
plt.title("Purchase Distribution")

plt.show()
```

🧪 [문제 5] 고액 주문 분석

✅ 문제

총 구매 금액이 200,000원 이상인 주문의 개수를 구하시오

```python
high_orders = df[df["TotalPrice"] >= 200000]

print("고액 주문 수:", len(high_orders))
```

🧪 [문제 5] 상품별 매출 분석

✅ 문제

1. ProductName별 평균 구매 금액을 구하시오.
2. 평균 구매 금액이 가장 높은 상위 5개 상품을 출력하시오.

```python
product_sales = df.groupby("ProductName")["TotalPrice"].mean()
print(product_sales)
```

```python
top_products = product_sales.sort_values(ascending=False).head(5)

print(top_products)
```
